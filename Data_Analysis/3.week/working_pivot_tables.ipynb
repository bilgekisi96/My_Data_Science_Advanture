```python
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Arthur', 'Michael', 'Jason', 'David', 'Thomas'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame({
    'id':[1,2,3,4,6],
   'Name': ['Mehmet', 'Terry', 'Joseph', 'Matt', 'Bruce'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
```

```python
pd.merge(left,right,on="id")  #neyin üzerinde birleştireceğiz
```

```python
pd.merge(left,right,on="id",how="outer") 
```

```python
pd.merge(left,right,on="id",how="right") #sağ tarafa göre yaparım diyor
```

```python
pd.merge(left,right,on=["id","subject_id"]) #ikisinde ortak olanı buldu
```

```python
pd.merge(left,right,on=["id","subject_id"])
```
