```python
from sqlalchemy import create_engine #bunu import etmeliyiz
```

```python
temp_db=create_engine("sqlite:///:memory:") #inisilaze etmeliyiz hangi sql servere bağlanacaksan yaziosun en basiti sqlite
#sabit veri vermiyoruz memory veriyoruz ramde tutar hafızayıda
temp_db.table_names()
```

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/World_population')
tables #bir tane tablo aldık
```

```python
pop=tables[5] #tablo 5 i çektik
pop
```

```python
pop.to_sql(name="populations",con=temp_db) #geçici database atabiliriz
```

```python
temp_db.table_names()
```

```python
pd.read_sql(sql="populations",con=temp_db)
```

```python
pd.read_sql_query(sql="select Country,Deaths from Populations",con=temp_db)
```
