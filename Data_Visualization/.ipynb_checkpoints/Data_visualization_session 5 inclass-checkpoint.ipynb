```python
# Basic Libraries
import pandas as pd
import numpy as np
# Libraries for visualizaton
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
titanic=sns.load_dataset("titanic")
titanic
```

```python
sns.catplot(x="pclass",y="survived",hue="sex",data=titanic,kind="bar",estimator=np.sum,row="embarked",col="alone")
```

```python
sns.catplot(x="pclass",y="survived",hue="sex",data=titanic,kind="bar",estimator=np.sum,col="sex")
```

```python
#####################    Pairgrid ########################
```

```python
g = sns.PairGrid(tips)
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.histplot)
```

```python
g=sns.PairGrid(tips)
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, color ="r");
```

```python
g = sns.PairGrid(tips, hue="sex", palette="viridis")
g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.distplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend()
```

```python
a = sns.pairplot(tips)
a.map_lower(sns.kdeplot, color='purple')
a.map_upper(sns.kdeplot, color='g');
```
