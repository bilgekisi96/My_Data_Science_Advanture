```python
import pandas as pd
import numpy as np
# Libraries for visualizaton
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
fig,ax=plt.subplots()
sns.countplot(x="day",data=tips)

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+0.5))
    ax.text(p.get_x()+0.1, p.get_height()-3,str(round(p.get_height()/tips.day.count(),2)) )     # yÃ¼zdelik ekledik
plt.show()
```

```python
ax=sns.countplot(x="day",data=tips,hue="sex")
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+0.3))
    ax.text(p.get_x()+0.1, p.get_height()-3,str(round(p.get_height()/tips.day.count(),2)) )     
    # yÃ¼zdelik ekledik
plt.show()
```

```python
df=pd.read_csv("StudentsPerformance.csv")
df
```
