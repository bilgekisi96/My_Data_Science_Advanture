```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x=np.array([10,11,12,15,19,22,25])
dicc,a=[(2*x,"*","-","r")],1 #,(x**3,"o","--","b"),(x**4,"."," ","y")],1
plt.figure(figsize=(15,10))
plt.suptitle("LineStyles & Markers & Markersize & Color",fontsize=16, y = 1.2)
for y,mark,lnsize,color,labl in dicc:
    plt.subplot(1,1,a)
    plt.plot(x, y, color, ls=lnsize, marker=mark ,markersize=15,alpha=0.9,label=labl)
    a+=1
plt.tight_layout();
plt.show()
```
